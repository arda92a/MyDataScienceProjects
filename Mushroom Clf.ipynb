```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
y=df["class"].values
X=df.iloc[:,1:23].values
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,shuffle=True,stratify=y)

svc=SVC()
svc.fit(xtrain,ytrain)
```

```python
yhead_svc=svc.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print("SVC accuracy on test:", accuracy_score(ytest,yhead_svc))
print("SVC accuracy on train:", accuracy_score(ytrain,svc.predict(xtrain)))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression(max_iter=500)
lg.fit(xtrain,ytrain)

yhead_lg=lg.predict(xtest)

print("Logistic regression accuracy on test:", accuracy_score(ytest,yhead_lg))
print("Logistic regression accuracy on train:", accuracy_score(ytrain,lg.predict(xtrain)))
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(xtrain,ytrain)

yhead_rf=clf.predict(xtest)

print("Random forest accuracy on test:", accuracy_score(ytest,yhead_rf))
print("Random forest accuracy on train:", accuracy_score(ytrain,clf.predict(xtrain)))

feature_importances=clf.feature_importances_
feature_name=df.iloc[:,1:23].columns
```

```python
feature_importance_df = pd.DataFrame({'Feature': feature_name, 'Importance': feature_importances})
```

```python
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
feature_importance_df
```
